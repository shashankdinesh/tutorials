```python
# OBTAINING COLUMN WISE NAN ELEMENT COUNT AND PERCENTAGE

# We'll consider that when more than 15% of the data is missing, we should delete the 
# corresponding variable and pretend it never existed. 
# This means that we will not try any trick to fill the missing data in these cases.

df_train=df_train_house_prices.copy(deep=True)
len(df_train) # it returns total number of rows in a dataframe 
total = df_train.isnull().sum().sort_values(ascending=False) # it gives column wise total NAN elements (series type)
total_nan =df_train.isnull().sum().sum() # it gives total NAN elements in all columns (series type)
df_train.count() # it gives count of non null element in a column (columnwise) (series type)
df_train.isnull().count() # it gives count of total elements(including null element) in a column (columnwise) (series type)
df_train.sum() # it add up all element in a column (if object type then element get concatenated) (series type)
df_train.isnull().count().sum() #  it gives count of total elements(including null element) in entire dataframe
df_train.count().sum() #  it gives count of total elements(excluding null element) in entire dataframe
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending=False)

missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent']) # it returns dataframe  

missing_data.head(20)


```
